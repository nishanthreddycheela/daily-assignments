```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer

# Load the dataset (Replace 'healthcare_data.csv' with the actual dataset file)
df = pd.read_csv('Day_15_Healthcare_Data.csv')

# Initial Exploratory Data Analysis (EDA)
print("Dataset Overview:")
print(df.head())
print("\nDataset Info:")
df.info()
print("\nMissing Values:")
print(df.isna().sum())

# Calculate percentage of missing values
missing_percent = df.isna().sum() / len(df) * 100
print("\nPercentage of Missing Values per Column:")
print(missing_percent)

# Visualizing missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isna(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()

# Imputation Techniques

# 1. Mean/Median/Mode Imputation
for col in df.select_dtypes(include=[np.number]).columns:
    df[col].fillna(df[col].median(), inplace=True)  # Using Median Imputation

for col in df.select_dtypes(include=["object"]).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)  # Using Mode Imputation

# 2. K-Nearest Neighbors (KNN) Imputation
knn_imputer = KNNImputer(n_neighbors=5)
df_knn = pd.DataFrame(knn_imputer.fit_transform(df.select_dtypes(include=[np.number])), columns=df.select_dtypes(include=[np.number]).columns)

# Evaluate the Effect of Imputation
print("\nStatistical Summary After Imputation:")
print(df.describe())

# Visualizing Impact of Imputation
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)
plt.title('Boxplot of Numerical Features After Imputation')
plt.show()

# Save the cleaned dataset
df.to_csv('cleaned_healthcare_data.csv', index=False)
print("\nCleaned dataset saved successfully!")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder

# Load the dataset (Replace 'healthcare_data.csv' with the actual dataset file)
df = pd.read_csv('Day_16_Healthcare_Data.csv')

# Initial Exploratory Data Analysis (EDA)
print("Dataset Overview:")
print(df.head())
print("\nDataset Info:")
df.info()
print("\nMissing Values:")
print(df.isna().sum())

# Handle Missing Data
for col in df.select_dtypes(include=[np.number]).columns:
    df[col].fillna(df[col].median(), inplace=True)  # Using Median Imputation

for col in df.select_dtypes(include=["object"]).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)  # Using Mode Imputation

# Detect and Handle Duplicates
duplicates = df.duplicated()
print(f"\nNumber of Duplicates: {duplicates.sum()}")
df = df.drop_duplicates()

# Detect and Handle Outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)
plt.title('Boxplot of Numerical Features Before Outlier Handling')
plt.show()

# Capping extreme values
for col in df.select_dtypes(include=[np.number]).columns:
    upper_limit = df[col].quantile(0.95)
    lower_limit = df[col].quantile(0.05)
    df[col] = np.clip(df[col], lower_limit, upper_limit)

# Standardize and Normalize Data
scaler = MinMaxScaler()
df[df.select_dtypes(include=[np.number]).columns] = scaler.fit_transform(df.select_dtypes(include=[np.number]))

# Convert categorical variables into numerical representations
encoder = LabelEncoder()
for col in df.select_dtypes(include=["object"]).columns:
    df[col] = encoder.fit_transform(df[col])

# Data Validation
print("\nFinal Dataset Info:")
df.info()
print("\nRemaining Missing Values:")
print(df.isna().sum())

# Save the cleaned dataset
df.to_csv('cleaned_healthcare_data.csv', index=False)
print("\nCleaned dataset saved successfully!")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import re

# Load the dataset (Replace 'Day_18_Tours_and_Travels.csv' with actual file name)
df = pd.read_csv('Day_18_Tours_and_Travels.csv')

# Initial Data Exploration
print("Dataset Overview:")
print(df.head())
print("\nDataset Info:")
df.info()
print("\nMissing Values:")
print(df.isna().sum())

# Handle Missing Values
num_imputer = SimpleImputer(strategy='median')
df['Customer_Age'] = num_imputer.fit_transform(df[['Customer_Age']])
df['Rating'] = num_imputer.fit_transform(df[['Rating']])

def handle_missing_text(text):
    return text if isinstance(text, str) else "No Review Provided"
df['Review_Text'] = df['Review_Text'].apply(handle_missing_text)

# Detect and Remove Duplicates
duplicates = df.duplicated()
print(f"\nNumber of Duplicates: {duplicates.sum()}")
df = df.drop_duplicates()

# Handle Inconsistent Data
# Ensure Rating is within 1-5
df['Rating'] = df['Rating'].clip(1, 5)

# Standardize Destination names (clean text for consistency)
def clean_text(text):
    return re.sub(r'[^a-zA-Z0-9 ]', '', text).strip().lower() if isinstance(text, str) else text
df['Destination'] = df['Destination'].apply(clean_text)

# Identify and Handle Outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Package_Price', 'Rating']])
plt.title('Boxplot of Package_Price and Rating Before Outlier Handling')
plt.show()

# Capping extreme values
for col in ['Package_Price', 'Rating']:
    upper_limit = df[col].quantile(0.95)
    lower_limit = df[col].quantile(0.05)
    df[col] = np.clip(df[col], lower_limit, upper_limit)

# Convert categorical data into numerical format for Destination column
encoder = LabelEncoder()
df['Destination'] = encoder.fit_transform(df['Destination'])

# Save the cleaned dataset
df.to_csv('cleaned_travel_reviews.csv', index=False)
print("\nCleaned dataset saved successfully!")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer

# Load the dataset
df = pd.read_csv("Day 19_E-Commerce_Data.csv")  # Replace with the correct filename

# Identify missing data
missing_values = df.isna().sum()
missing_percentage = (df.isna().sum() / len(df)) * 100

print("Missing Values:")
print(missing_values)
print("\nMissing Percentage:")
print(missing_percentage)

# Visualize missing data patterns
plt.figure(figsize=(10,6))
sns.heatmap(df.isna(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Missing Data Heatmap")
plt.show()

# Handle missing values
# 1. Mean/Median imputation for numerical columns
numerical_cols = ["Product_Price"]  # Add other numerical columns if needed
for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)  # Median imputation

# 2. Mode imputation for categorical columns
categorical_cols = ["Product_Category"]  # Add other categorical columns if needed
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# 3. Forward fill for date-related fields
date_cols = ["Order_Date"]  # Add other date columns if needed
for col in date_cols:
    df[col].fillna(method='ffill', inplace=True)

# 4. KNN Imputation for complex cases
knn_imputer = KNNImputer(n_neighbors=5)
df_knn = pd.DataFrame(knn_imputer.fit_transform(df.select_dtypes(include=[np.number])), columns=df.select_dtypes(include=[np.number]).columns)
df[df_knn.columns] = df_knn

# Evaluate the impact
print("Summary Statistics Before Imputation:")
print(df.describe())

# Visualization after imputation
plt.figure(figsize=(10,6))
sns.boxplot(data=df[numerical_cols])
plt.title("Boxplot of Imputed Numerical Data")
plt.show()

# Save cleaned dataset
df.to_csv("ecommerce_orders_cleaned.csv", index=False)

print("Data cleaning completed successfully!")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Day 19_E-Commerce_Data.csv')

# 1. Handle Missing Values
# Impute missing Rating with median
df['Rating'] = df['Rating'].fillna(df['Rating'].median())

# Impute missing Customer_Age with median
df['Customer_Age'] = df['Customer_Age'].fillna(df['Customer_Age'].median())

# Fill missing Review_Text with placeholder
df['Review_Text'] = df['Review_Text'].fillna('No review')

# 2. Detect and Remove Duplicates
df = df.drop_duplicates(subset=['Review_Text'], keep='first')

# 3. Handle Inconsistent Data
# Ensure Rating is between 1 and 5
df['Rating'] = df['Rating'].apply(lambda x: max(1, min(x, 5)))

# Standardize Product_Category (correct spelling issues)
df['Product_Category'] = df['Product_Category'].replace({
    'elctronics': 'electronics',
    'beauty': 'beauty & personal care',
    # Add more corrections as needed
})

# 4. Identify and Handle Outliers
# Boxplot for Product_Price
sns.boxplot(df['Product_Price'])
plt.show()

# Boxplot for Rating
sns.boxplot(df['Rating'])
plt.show()

# Remove outliers from Product_Price (for example, values above 95th percentile)
price_threshold = df['Product_Price'].quantile(0.95)
df = df[df['Product_Price'] <= price_threshold]

# Remove outliers from Rating (values above 5)
df = df[df['Rating'] <= 5]

# 5. Prepare Data for Analysis
# One-hot encode categorical columns
df = pd.get_dummies(df, columns=['Product_Category'], drop_first=True)

# 6. Save the Cleaned Dataset
df.to_csv('cleaned_customer_reviews.csv', index=False)
```
