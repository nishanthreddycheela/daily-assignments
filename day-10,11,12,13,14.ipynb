```python

# Load the CSV file
df = pd.read_csv('Day_10_banking_data.csv')

# Task 1: Filter out all rows where the Transaction_Amount is greater than 2000
filtered_transaction_amount = df[df['Transaction_Amount'] <= 2000]
print("Transactions with Transaction_Amount <= 2000:")
print(filtered_transaction_amount)

# Task 2: Find all rows where the Transaction_Type is "Loan Payment" and Account_Balance is greater than 5000
filtered_loan_payment = df[(df['Transaction_Type'] == 'Loan Payment') & (df['Account_Balance'] > 5000)]
print("\nTransactions with Transaction_Type 'Loan Payment' and Account_Balance > 5000:")
print(filtered_loan_payment)

# Task 3: Extract transactions made in the 'Uptown' branch
filtered_uptown_branch = df[df['Branch'] == 'Uptown']
print("\nTransactions made in 'Uptown' branch:")
print(filtered_uptown_branch)
```

```python
import pandas as pd

# Assuming your CSV file is named 'banking_data.csv'
df = pd.read_csv('Day_11_banking_data.csv')

# Strip any extra spaces from column names if needed
df.columns = df.columns.str.strip()

# Task 1: Sort the dataset by Account_Balance in descending order and display the first 10 rows
sorted_by_balance = df.sort_values(by='Account_Balance', ascending=False)
print("Top 10 rows sorted by Account_Balance in descending order:")
print(sorted_by_balance.head(10))

# Task 2: Create a ranking column for Transaction_Amount within each Branch
df['Transaction_Rank'] = df.groupby('Branch')['Transaction_Amount'].rank(ascending=False)

# Display the relevant columns along with the new ranking
print("\nData with ranking of Transaction_Amount within each Branch:")
print(df[['Date', 'Account_Type', 'Branch', 'Transaction_Type', 'Transaction_Amount', 'Transaction_Rank']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Day_11_banking_data.csv')

# Strip any extra spaces from column names
df.columns = df.columns.str.strip()

# Task 1: Plot the total sum of Transaction_Amount per Account_Type using a bar plot
total_transaction_by_account = df.groupby('Account_Type')['Transaction_Amount'].sum()

# Create a bar plot
plt.figure(figsize=(8, 6))
total_transaction_by_account.plot(kind='bar', color='skyblue')
plt.title('Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Task 2: Create a pie chart to show the percentage of transactions per Branch
transactions_per_branch = df['Branch'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
transactions_per_branch.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')  # Hide the y-label as it's not needed for a pie chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the CSV file
df = pd.read_csv('Day_13_Pharma_data.csv')  

# Data Cleaning: Check for missing values and duplicates
print("Missing values in the dataset:\n", df.isnull().sum())
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Visualization 1: Bar plot showing total sales per region
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title('Total Sales Per Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()

# Visualization 2: Scatter plot to visualize the relationship between Marketing_Spend and Sales
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=df)
plt.title('Marketing Spend vs Sales')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.show()

# Visualization 3: Boxplot comparing drug effectiveness across different age groups
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age_Group', y='Effectiveness', data=df)
plt.title('Drug Effectiveness Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Effectiveness')
plt.show()

# Visualization 4: Line plot showing the sales trend for each product over different trial periods
plt.figure(figsize=(8, 6))
sns.lineplot(x='Trial_Period', y='Sales', hue='Product_Name', data=df, marker='o')
plt.title('Sales Trend for Each Product Over Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Sales')
plt.show()

# Visualization 5: Heatmap showing correlation between Sales, Marketing_Spend, and Effectiveness
correlation_data = df[['Sales', 'Marketing_Spend', 'Effectiveness']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Between Sales, Marketing Spend, and Effectiveness')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the CSV file
df = pd.read_csv('Day_13_Pharma_data.csv')  # Replace 'your_dataset.csv' with the actual file name

# Data Cleaning: Check for missing values and duplicates
print("Missing values in the dataset:\n", df.isnull().sum())
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Visualization 1: Bar plot comparing average Effectiveness for each drug across different regions
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Effectiveness', hue='Product_Name', data=df, ci=None)
plt.title('Average Effectiveness of Drugs by Region')
plt.xlabel('Region')
plt.ylabel('Average Effectiveness')
plt.show()

# Visualization 2: Violin plot to show the distribution of Effectiveness and Side_Effects for each product
plt.figure(figsize=(10, 6))
sns.violinplot(x='Product_Name', y='Effectiveness', data=df, inner="quart")
plt.title('Distribution of Effectiveness for Each Product')
plt.xlabel('Product Name')
plt.ylabel('Effectiveness')
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='Product_Name', y='Side_Effects', data=df, inner="quart")
plt.title('Distribution of Side Effects for Each Product')
plt.xlabel('Product Name')
plt.ylabel('Side Effects')
plt.show()

# Visualization 3: Pairplot to explore relationships between Effectiveness, Side_Effects, and Marketing_Spend
sns.pairplot(df[['Effectiveness', 'Side_Effects', 'Marketing_Spend']])
plt.suptitle('Pairplot: Effectiveness, Side Effects, and Marketing Spend', y=1.02)
plt.show()

# Visualization 4: Boxplot comparing Effectiveness for different trial periods
plt.figure(figsize=(8, 6))
sns.boxplot(x='Trial_Period', y='Effectiveness', data=df)
plt.title('Effectiveness Across Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.show()

# Visualization 5: Regression plot to analyze how Marketing_Spend affects drug Effectiveness
plt.figure(figsize=(8, 6))
sns.regplot(x='Marketing_Spend', y='Effectiveness', data=df, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Effectiveness vs Marketing Spend')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.show()

```
